```python
import numpy as np
import pandas as pd
```

```python
input_values = np.array([[0,0],[0,1],[1,1],[1,0]])
input_values.shape
input_values
```

```python
output = np.array([0,1,1,0])
output.reshape(4,1)
output
```

```python
weights = np.array([[0.1],[0.2]])
weights
```

```python
bias = 0.3
```

```python
def sigmoid_fun(x):
  return 1/(1 + np.exp(-x))
```

```python
def der(x):
  return sigmoid_fun(x)*(1-sigmoid_fun(x))
```

```python
for i in range(10000):
  input_arr = input_values

  weighed_sum = np.dot(input_arr,weights) + bias
  first_output = sigmoid_fun(weighed_sum)
  error = first_output - output
  total_error = np.square(np.subtract(first_output,output)).mean()

  first_der = error
  second_der = der(first_output)
  derivate = first_der*second_der

  t_input = input_values.T
  final_derivative = np.dot(t_input,derivate)

  weights = weights - 0.05*final_derivative

  for i in derivate:
    bias = bias - 0.05*i
  
  print(weights)
  print(bias)
```
